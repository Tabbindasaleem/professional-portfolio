```python
import pandas as pd
df=pd.read_csv("amazon.csv")
print(df)
```

```python
df["product_id"].unique()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
for col in ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']:
    df[col] = df[col].astype(str) \
        .str.replace('‚Çπ', '') \
        .str.replace('%', '') \
        .str.replace(',', '') \
        .str.replace('nan', '0')

    df[col] = pd.to_numeric(
        df[col].str.replace('[^0-9.]', '', regex=True), errors='coerce'
    ).fillna(0)
```

```python
df.describe()
```

```python
# foe finding thw top products
top_products=df["product_name"].value_counts().head(10)
print(top_products)
```

```python
print("Overall Discount Statistics:")
print(df['discount_percentage'].describe())
```

```python
top_discounts = df.sort_values(by='discount_percentage', ascending=False)
print(top_discounts[['product_name', 'actual_price', 'discounted_price', 'discount_percentage']].head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df, x='discount_percentage', y='rating')
plt.title('Discount vs Product Rating')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='actual_price', y='discount_percentage')
plt.title('Actual Price vs Discount Percentage')
plt.show()
```

```python
df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Avg Discount % by Category')
plt.ylabel('Discount %')
plt.show()
```

```python
high_rated = df[df['rating_count'] > 100]  # Filter to avoid bias from small counts
top_rated = high_rated.sort_values(by='rating', ascending=False).head(10)
print(top_rated[['product_name', 'rating', 'rating_count']])
```

```python
df['price_range'] = pd.cut(df['actual_price'], bins=[0, 500, 1000, 2000, 5000, 10000, df['actual_price'].max()],
                           labels=['0-500', '500-1000', '1k-2k', '2k-5k', '5k-10k', '10k+'])

discount_by_range = df.groupby('price_range')['discount_percentage'].mean()

discount_by_range.plot(kind='bar', title='Avg Discount by Price Range')
```

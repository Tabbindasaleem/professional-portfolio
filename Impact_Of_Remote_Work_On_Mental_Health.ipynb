```python
from scipy.stats import ttest_ind, f_oneway, chi2_contingency
onsite_balance = df[df['Work_Location'] == 'Onsite']['Work_Life_Balance_Rating']
remote_balance = df[df['Work_Location'] == 'Remote']['Work_Life_Balance_Rating']
hybrid_balance = df[df['Work_Location'] == 'Hybrid']['Work_Life_Balance_Rating']

# Perform ANOVA for Work-Life Balance Rating across different Work Locations
anova_result = f_oneway(onsite_balance, remote_balance, hybrid_balance)
print("ANOVA Result:", anova_result)

# Perform T-tests between each pair of Work Locations
t_test_onsite_remote = ttest_ind(onsite_balance, remote_balance, equal_var=False)
t_test_onsite_hybrid = ttest_ind(onsite_balance, hybrid_balance, equal_var=False)
t_test_remote_hybrid = ttest_ind(remote_balance, hybrid_balance, equal_var=False)

print("T-test Onsite vs Remote:", t_test_onsite_remote)
print("T-test Onsite vs Hybrid:", t_test_onsite_hybrid)
print("T-test Remote vs Hybrid:", t_test_remote_hybrid)

# Chi-square test for association between Mental Health Condition and Work Location
contingency_table = pd.crosstab(df['Mental_Health_Condition'], df['Work_Location'])
chi2_result = chi2_contingency(contingency_table)
print("Chi-square Result:", chi2_result)
```

```python
#FINAL REULTS
#ANOVA Test for Work-Life Balance Rating Across Work Locations
#This suggests no statistically significant difference in work-life balance ratings among employees working onsite,
#remotely, or in hybrid settings. In other words, work location does not have a significant impact on the work-life balance rating.
#Overall, the results suggest that neither work location nor work conditions in this dataset have a statistically significant impact on 
#work-life balance ratings or mental health conditions.
```
